## Task 2 Efficiently clean the data

Much of the USGS data is not needed, or bunches of data may be missing. 

Make a function called ``clean_df()`` that takes in a dataframe, null columns, and a specific state and conducts the following actions:

* Removes the unwanted columns (e.g., create a variable of a list of undesireable columns such as 'variable', 'measurement_unit', 'qualifiers', 'series')
* Adds a column for state and fills the column with the states name
* Sets the date column to a pandaas datetime object (e.g., pd.to_datetime())


Apply the cleaning function to each dictionary of dataframes

```python
# create a function called clean_df() that takes in a dataframe and does the following:
# removes variable, measurement_unit, qualifiers, and series columns
# converts the Datetime column to datetime format and sets it as the index
# add a column called "state" that contains the state abbreviation (the title of the folder the data is in)
def clean_df(df, state):
    df = df.drop(columns=["variable", "measurement_unit", "qualifiers", "series"])
    df["Datetime"] = pd.to_datetime(df["Datetime"]) 
    df["State"] = state
    return df


```

```python
# clean ID data
ID_dict = {key: clean_df(df, "Idaho") for key, df in ID_dict.items()}
# clean UT data
UT_dict = {key: clean_df(df, "Utah") for key, df in UT_dict.items()}
# clean WY data
WY_dict = {key: clean_df(df, "Wyoming") for key, df in WY_dict.items()}
```
